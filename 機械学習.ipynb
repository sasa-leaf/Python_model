# 線形回帰モデル

予測したい値（**目的変数**）を他の値（**説明変数**）で予測する分析。<br>
データの特徴の傾向を**直線**に表し、何かしらの**連続する数値**を予測する。<br>
株価の予測、家賃の予想、売上の予測、社会問題（数値がある）解決の方法の研究など。

## 単回帰分析

- $y=ax+b$

## 重回帰分析

複数の説明変数Xを用いて目的変数Yを表す。<br>
- $Y=w_0+w_1x_1+w_2x_2\cdots+w_nx_n$

最小二乗法では、サンプル点と重回帰式に対する残差二乗和（SSE）を最小化する係数$w_0$〜$w_n$を推定する。<br>
最小二乗法の指標としては、RMSE(平均平方二乗誤差)またはMAE(平均絶対誤差)が用いられる。<br>
いずれも、値が小さい方が誤差が小さいということ。

$$RMSE=\sum_{i=1}^{n} (y_i-\hat{y}_i)^2$$

$$MAE=\sum_{i=1}^{n} |y_i-\hat{y}_i|$$

$n$ : サンプル総数<br>
$y_i$ : 実測値<br>
$\hat{y}_i$ : 予測値

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x3,y)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter3D(x1, x2, y)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")

mesh_x1 = np.arange(x1.min()[0], x1.max()[0], (x1.max()[0]-x1.min()[0])/10)
mesh_x2 = np.arange(x2.min()[0], x2.max()[0], (x2.max()[0]-x2.min()[0])/10)
mesh_x1, mesh_x2 = np.meshgrid(mesh_x1, mesh_x2)
mesh_y = model.coef_[0][0] * mesh_x1 + model.coef_[0][1] * mesh_x2 + model.intercept_[0]
ax.plot_wireframe(mesh_x1, mesh_x2, mesh_y, color='red')
plt.show()
```

## 正則化

「正則化」とは、過学習を防いで汎化性能（未知のデータへの対応能力）を高めるためのテクニックの一つ。<br>
**コスト関数**（モデルの悪さを表す）に**正則化項**を付けることでモデルの形が複雑になりすぎないように調整しようとするもの。

### リッジ回帰

最小二乗コスト関数に、学習した重みの二乗（**L2正則化項**）を加えたもの。<br>
重みが完全に０にはならない性質があるため、説明変数が非常に多い時にはモデルの解釈が複雑になるという欠点がある。

$$E_Ridge=\sum_{i=1}^{n} (y_i-\hat{y}_i)^2+{\frac{1}{2}}λ\sum_{k=1}^{K}w_k^2$$

### Lasso回帰

最小二乗コスト関数に、学習した重み（**L1正則化項**）を加えたもの。<br><br>
不要と判断される説明変数の重みが0になる性質があり、モデル構築においていくつかの特徴量が完全に無視される。<br>
モデルを解釈がしやすく（スパースな解が求められる）なるため、どの説明変数が目的変数へ作用する度合いが強いのかが分かりやすい。<br><br>
モデルに取り込むことができる特徴量数に限りがある。

$$E_Lasso=\sum_{i=1}^{n} (y_i-\hat{y}_i)^2 +λ\sum_{i=1}^{n}|w_k|$$
