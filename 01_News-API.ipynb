# NewsAPIで日本語のニュースを取得する

```python
pip install newsapi-python
```

```python
from newsapi import NewsApiClient
from pprint import pprint
```

```python
API_KEY = 'each person input'
word = 'アップル'
language = None #'ja'がないのでNoneとする
country = 'jp' #日本国内だから'jp'とする
```

## タイトルで取得

```python
# Init
newsapi = NewsApiClient(api_key=API_KEY) #各自のもの

# /v2/top-headlines
top_headlines = newsapi.get_top_headlines(q=word, language=language, country=country)
```

```python
pprint(top_headlines)
```

題名と説明のみ

```python
top = top_headlines['articles']
last = len(top)

for i in range(0,last):
    article = top[i]
    print("題名:",article['title'])
    print("説明:",article['description'])
    print()
```

文字のみ

```python
top = top_headlines['articles']
last = len(top)

list1 = []
for i in range(0,last):
    article = top[i]
    a = article['title']
    b = article['description']
    list1.append(a)
    list1.append(b)
    
x = ''.join(list1)
print(x)
```

## 本文で取得

```python
# Init
newsapi = NewsApiClient(api_key=API_KEY)

# /v2/everything
all_articles = newsapi.get_everything(q=word, language=language)
```

```python
pprint(all_articles)
```

題名と説明のみ

```python
all_ = all_articles['articles']
last = len(all_)

for i in range(0,last):
    article = all_[i]
    print("題名:",article['title'])
    print("説明:",article['description'])
    print()
```

文字のみ

```python
all_ = all_articles['articles']
last = len(all_)

list2 = []
for i in range(0,last):
    article = all_[i]
    c = article['title']
    d = article['description']
    list2.append(c)
    list2.append(d)

y = ''.join(list2)
print(y)    
```

# テキストマイニング

## テキストファイルにする

```python
file = open('output/'+word+'.txt', mode='w')
z = x + y
file.write(z)
file.close()
```

## テキストマイニング

```python
read_file_name = 'output/'+word+'.txt'
create_file_name = 'output/'+word+'.png'
font = "/System/Library/Fonts/ヒラギノ角ゴシック W5.ttc"
```

```python
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

with open(read_file_name,encoding='utf-8')as _f:
    text = _f.read()
    
wordcloud = WordCloud(background_color="white",font_path=font,width=800,height=600).generate(text)
wordcloud.to_file(create_file_name)
im = Image.open(create_file_name)
plt.figure(figsize=(16,12))
plt.imshow(im)
plt.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.tick_params(bottom=False,left=False,right=False,top=False)
plt.show()
```
